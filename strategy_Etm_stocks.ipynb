```python
price_df, volume_df = YF.get_yf_data(tickers, start_date, end_date)
```

```python
price_df = np.round(price_df, 2)
price_df.head()
```

```python
price_df.sort_index()
```

```python
min_df = price_df.agg([min]).T.reset_index()
min_df
```

```python
max_df = price_df.agg([max]).T.reset_index()
max_df
```

```python
min_df = pd.merge(min_df, price_df.idxmin().to_frame(name="min_date").reset_index(), on="Ticker")
min_df
```

```python
max_df = pd.merge(max_df, price_df.idxmax().to_frame(name="max_date").reset_index(), on="Ticker")
max_df
```

```python
cmp_df = price_df.iloc[-1].to_frame(name="cmp").reset_index()
cmp_df
```

```python
final_df = pd.merge(max_df, min_df, on="Ticker")
final_df = pd.merge(final_df, cmp_df, on="Ticker")
final_df
```

```python
final_df['current_date'] = end_date
final_df['current_date'] = pd.to_datetime(final_df['current_date'])
```

```python
price_df.tail()
```

```python
final_df['delta_days']= (final_df.current_date - final_df.min_date).dt.days
final_df.head()
```

```python
final_df['date_score'] = 1/(final_df['delta_days']+1)
```

```python
df.head()
```

```python
def get_ret_from_52low(s):
  #print(s.min_date)
  x = price_df[price_df.index >=  s.min_date][s.Ticker].ffill().pct_change().dropna()
  #return round(np.power(np.prod(1+x), 1/len(x)) -1, 2)*100
  return x.mean()
```

```python
final_df['52L_mean_return'] = final_df.apply(get_ret_from_52low, axis=1)
```

```python
final_df
```

```python
final_df = pd.merge(final_df, df[['Ticker', 'pe', 'ipe', 'peg']], on='Ticker')
```

```python
final_df['pescore'] = final_df.ipe/final_df.pe
```

```python
final_df['pegscore'] = 1/final_df.peg
```

```python
final_df.rename(columns={'Ticker':'codes'}, inplace=True)
```

```python
final_df
```

```python
final_df.sort_values(by='pegscore', ascending=False)
```

```python
N = 2
trades = pd.DataFrame()
'''
S_dict = {

         'etm52ldays' : 'date_score',
         'etmomunpefunda' : 'pescore',
         'etmpeg' : 'pegscore'
}
'''
S_dict = {

         'fiipe' : 'pescore',
         'fiipeg' : 'pegscore'
}

#codes	Action	strategy	weight

for strategy in S_dict:
    temp = final_df.sort_values(by=S_dict[strategy], ascending=False)
    temp = temp.head(N)
    temp['weight'] = temp[S_dict[strategy]]/temp[S_dict[strategy]].sum()
    temp['strategy'] = strategy
    temp['Action'] = 'credit'
    print(temp[['codes', 'Action', 'strategy', 'weight']])
    temp[['codes', 'Action', 'strategy', 'weight']].to_csv(f'/content/tradebook_{strategy}.csv', index=False)
```
